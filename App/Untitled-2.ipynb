```python
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python

# Définir les colonnes de caractéristiques (exclure identifiants)
feature_cols = [col for col in df.columns if col not in ['Designation', 'Machines']]

# Fonction pour extraire les caractéristiques d'une ligne
def get_state_features(row):
    return tuple(row[feature_cols].values)  # Convertir en tuple pour la table Q

# Initialiser la table Q
Q = defaultdict(lambda: np.zeros(len(df)))  # États x Actions

# Paramètres RL
alpha, gamma, epsilon = 0.1, 0.9, 0.1  # Taux d'apprentissage, facteur de réduction, exploration

# Fonction de récompense
def get_reward(current_features, next_features):
    reward = -next_features['Couverture']  # Priorité à faible couverture
    diff = np.sum((np.array(current_features) - np.array(next_features)) ** 2) ** 0.5  # Distance euclidienne
    reward -= diff * 0.1  # Pénalité pour différence de format
    return reward

# Algorithme de planification avec RL
def plan_fabrication(machine_df):
    if machine_df.empty:
        return []

    states = [get_state_features(row) for _, row in machine_df.iterrows()]
    actions = machine_df.index.tolist()
    order = []

    current_state = states[0]
    current_idx = actions[0]
    order.append(machine_df.loc[current_idx, 'Designation'])

    for _ in range(len(machine_df) - 1):
        if np.random.random() < epsilon:
            next_idx = np.random.choice([a for a in actions if a != current_idx])
        else:
            q_values = [Q[current_state][a] for a in actions if a != current_idx]
            next_idx = actions[np.argmax(q_values)] if q_values else np.random.choice(actions)

        next_state = states[actions.index(next_idx)]
        reward = get_reward(
            machine_df.loc[current_idx, feature_cols],
            machine_df.loc[next_idx, feature_cols]
        )

        # Mise à jour de Q
        old_value = Q[current_state][current_idx]
        next_max = np.max([Q[next_state][a] for a in actions])
        Q[current_state][current_idx] = old_value + alpha * (reward + gamma * next_max - old_value)

        order.append(machine_df.loc[next_idx, 'Designation'])
        current_state = next_state
        current_idx = next_idx

    return order

# Exécuter par machine
machines = df['Machines'].unique()
optimized_orders = {}
for machine in machines:
    machine_df = df[df['Machines'] == machine].copy()
    optimized_orders[machine] = plan_fabrication(machine_df)

# Afficher les résultats
for machine, order in optimized_orders.items():
    print(f"Machines {machine}: {', '.join(order)}")

# Optionnel : Sauvegarder
output_df = pd.DataFrame({
    "Machines": [machine for machine in optimized_orders.keys() for _ in optimized_orders[machine]],
    "Ordre_Fabrication": [prod for order in optimized_orders.values() for prod in order]
})
output_df.to_excel("plan_fabrication.xlsx", index=False)
```
